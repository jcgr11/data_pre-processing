```python
#import necessary packages
import pandas as pd
import yfinance as yf
import cufflinks as cf
from plotly.offline import iplot, init_notebook_mode
init_notebook_mode(connected =True)
cf.go_offline()
```

```python
#Download price data
axp = yf.download ("AXP", 
                   start = "2022-03-31", 
                   end = "2023-03-31", 
                   progress = False, 
                   auto_adjust = True)
```

```python
#call axp dataframe to cufflinks quantfig, set title, legend position (legend is automatically set by quantfig function), and ticker )
qf = cf.QuantFig(
    axp, title = "AXP Daily Price Movement, SMA, EMA, & Volume", legend = "top", name = "AXP stock price"
)
```

```python
# Plot daily volume, simple moving average & exponential moving average based on average number of trading days in a month, and line colors
qf.add_volume()
qf.add_sma(periods = 20, column = "Close", color = "red")
qf.add_ema(periods = 20, color = "green")

# Save the plot as a static image
qf.iplot(asFigure=True)
```
